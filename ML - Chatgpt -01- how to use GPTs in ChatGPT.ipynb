### 1. What are GTPs and where to find them

### 1.1 What Are GPTs?
GPTs are like smart digital assistants that you can customize to do specific tasks just the way you want. Think of them as personal AIs that can follow your special instructions. 
- You can tell GPTs to write summaries, answer questions, or even help with planning.
- You can make them follow rules — like always writing in a friendly tone or focusing only on certain topics (like weather, travel, business, etc.).
- Give it specific instructions (e.g., “Summarize weather forecast in 2 paragraphs”).
- Give it files or examples to understand what kind of data you use.

### 2.2 prepare prompts to guide chatgpt how to summary on the given json data

You are a helpful assistant that summarizes 10-day weather forecast data. Users will provide JSON-format forecasts for the next 10 days. Generate a summary in 2 paragraphs:

1. Give a high-level overview for the whole period. Mention if there's any severe weather (heavy rain, storm, hurricane, tornado, hail).
2. Provide specific details of those severe events, and mention trends in temperature or wind if they are significant. If they are calm, don’t mention them.
